## Scipy tutorial
### Blurring Images
    讲解scipy的使用和对图像的操作变换

```python

from PIL import Image
from numpy import *
from scipy.ndimage import filters
from pylab import *
import warnings
import sys



sys.path.append('../source_data')
warnings.filterwarnings("always")

empire_img_path = '../source_data/empire.jpg'
imshow(array(Image.open(empire_img_path)))
# print(sys.path)
```

## 对灰度图像做高斯模糊变换

```python
im = array(Image.open('../source_data/empire.jpg').convert('L'))
im2 = filters.gaussian_filter(im, 5)
gray()
figure()
subplot(1, 2, 1)
imshow(im)
title('origin')
subplot(1, 2, 2)
imshow(im2)
title('gaussian')
show()
```

## 对彩色图像进行高斯模糊变换

```python
im = array(Image.open('../source_data/empire.jpg'))
im2 = zeros(im.shape)
for i in range(3):
#     print("----------")
    im2[:, :, i] = filters.gaussian_filter(im[:, :, i], 5)
    im2 = uint8(im2)
    imshow(im2)
```

## Image derivatives
### sobel 与 gaussian 滤波器做对比

```python
im = array(Image.open(empire_img_path).convert('L'))
gray()
imx = zeros(im.shape)
filters.sobel(im, 1, imx)
imy = zeros(im.shape)
filters.sobel(im, 0, imy)
magnitude = sqrt(imx**2 + imy**2)
figure()
subplot(1, 4, 1);imshow(im);title('origin')
subplot(1, 4, 2);imshow(uint8(imx));title('x-derivative')
subplot(1, 4, 3);imshow(uint8(imy));title('y-derivative')
subplot(1, 4, 4);imshow(array(magnitude, 'uint8'));title('gradient magnitude')


sigma = 5
imx = zeros(im.shape)
filters.gaussian_filter(im, (sigma, sigma), (0, 1), imx)
imy = zeros(im.shape)
sigma = 1
filters.gaussian_filter(im, (sigma, sigma), (0, 1), imy)

figure()
subplot(1,2, 1); imshow(imx); title('guasissian - x')
subplot(1, 2, 2); imshow(imy); title('guasissian - y')
# show()
```
